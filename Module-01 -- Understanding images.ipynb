Import libraries.

```python
from PIL import Image
import matplotlib
import cv2
```

Read image through PIL.

```python
img = Image.open("./data/test.jpg")
img
```

Read through matplotlib.

```python
img_mat = matplotlib.image.imread("./data/test.jpg")
img_mat
```

Reads through OpenCV.

```python
img_cv = cv2.imread("./data/test.jpg")
img_cv
```
