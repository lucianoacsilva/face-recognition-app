# Image resizing

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("./data/test.jpg")
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
print(gray_img.shape)
plt.imshow(gray_img, cmap= "gray")
plt.show()
```

Shrink image.

```python
img_resized = cv2.resize(gray_img, (500, 400), cv2.INTER_AREA)
img_resized.shape
plt.imshow(img_resized, cmap= "gray")
```

Enlarge image.

```python
img_enlarged = cv2.resize(gray_img, (2000, 1600), cv2.INTER_CUBIC)
plt.imshow(img_enlarged, cmap= "gray")
```
