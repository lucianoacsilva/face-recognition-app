# Image cropping

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from glob import glob
```

```python
female_path = glob("./data/female/*.jpg")
male_path = glob("./data/male/*.jpg")
```

```python
len(female_path), len(male_path)
```

Reads image.

```python
path = female_path[0]
img = cv2.imread(path)
img.shape
```

```python
plt.imshow(img)
plt.show()
```

Convert to gray scale.

```python
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image, cmap= "gray")
```

Applies haar cascade classifier.

```python
haar = cv2.CascadeClassifier("./model/haarcascade_frontalface_default.xml")
faces = haar.detectMultiScale(gray_image, 1.5, 5)
faces
```

Draws cropping box.

```python
for x, y, width, height in faces:
    cv2.rectangle(img, (x, y), (x + width, y + height), (0, 255, 0), 2)
    
plt.imshow(img)
```

Crops image.

```python
crop_img = img[y: y + height, x: x + width]
plt.imshow(crop_img)
cv2.imwrite("f_01.png", crop_img)
```

Apply to all images.

```python
def crop_image(img_path, gender, i):
    img = cv2.imread(img_path)
    gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = haar.detectMultiScale(gray_image, 1.5, 5)

    
    for x, y, width, height in faces:
        crop_img = img[y: y + height, x: x + width]
        
        if (gender == "male"):
            cv2.imwrite('./data/crop/male_crop/{}_{}.png'.format(gender, i), crop_img)
            
        else:
            cv2.imwrite('./data/crop/female_crop/{}_{}.png'.format(gender, i), crop_img)
        
```

Crop female images.

```python
for i, path in enumerate(female_path):

    try:
        crop_image(path, "female", i)
        print("INFO: {}/{} processed sucessfully".format(i, len(female_path)))
        
    except:
    
        print("INFO: {}/{} cannot be processed".format(i, len(female_path)))
```

Crop male images.

```python
for i, path in enumerate(male_path):

    try:
        crop_image(path, "male", i)
        print("INFO: {}/{} processed sucessfully".format(i, len(female_path)))
        
    except:
    
        print("INFO: {}/{} cannot be processed".format(i, len(female_path)))
```
