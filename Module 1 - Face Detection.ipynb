# Face Detection

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Reads image
img = cv2.imread("./data/male_000281.jpg")

# Converts image to grayscale
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Applies haar cascade
haar = cv2.CascadeClassifier("./data/haarcascade_frontalface_default.xml")
faces = haar.detectMultiScale(gray_img, 1.3, 5)
faces
```

```python
cv2.rectangle(img, (faces[0][0], faces[0][1]), (faces[0][0] + faces[0][2], faces[0][1] + faces[0][3]), (0, 255, 0), 5)
plt.imshow(img)
```

```python
cv2.imshow("object_detect",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Crops face.

```python
crop = img[faces[0][1]: faces[0][1] + faces[0][3], faces[0][0]: faces[0][0] + faces[0][2]]
#crop = img[94: 94 + 261, 154: 154 + 261]
plt.imshow(crop)
```

Save in file.

```python
cv2.imwrite("male_01_classified.png", crop)
```
